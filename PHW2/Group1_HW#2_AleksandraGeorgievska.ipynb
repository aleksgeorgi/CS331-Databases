Proposition: Return all customers, and for each return a Yes/No value
depending on whether the customer placed an order on Feb 12, 2016


```sql
USE Northwinds2022TSQLV7

Select DISTINCT C.CustomerId, C.CustomerCompanyName, --O.OrderId, O.OrderDate,
    CASE WHEN O.OrderId IS NOT NULL THEN 'Yes' ELSE 'No' END AS HasOrderOn20160212
FROM Sales.[Customer] AS C
    LEFT OUTER JOIN Sales.[Order] AS O
    ON O.CustomerId = C.CustomerId
        AND O.orderDate = '20160212'
-- WHERE C.CustomerId = 45 or C.CustomerId = 48
ORDER BY CustomerId
```

### Using Outer Joins in a Multi-Join Query

proposition: For each customer that has made an order (1 or more), obtain the product IDs and quantity of each product that makes up each order placed. Use a Left Outer Join first on the Customer and Order tables, and then use the results of that join to perform an Inner Join on the OrderDetails table  


```sql
USE Northwinds2022TSQLV7

SELECT C.CustomerId, O.OrderId, OD.ProductId, OD.Quantity
FROM Sales.[Customer] AS C
  LEFT OUTER JOIN Sales.[Order] AS O
    ON C.CustomerId = O.CustomerId
  INNER JOIN Sales.[OrderDetail] AS OD
    ON O.orderid = OD.orderid
ORDER BY C.CustomerId, O.OrderId, OD.Quantity DESC, OD.ProductId
```

### Option 1: use outer join all along

proposition: For each customer that has made an order (1 or more), obtain the product IDs and quantity of each product that makes up each order placed. Use an Outer Join first on the Customer and Order tables, and then use the results of that join to perform another Outer Join on the OrderDetail table. 

- Note: this option returns 2 more results than the previous query strategy, and those results are NULLs for the Customers that never placed an order. This is a result of using only outer joins, which preserves NULLS

```sql
Use Northwinds2022TSQLV7

SELECT C.CustomerId, O.OrderId, OD.ProductId, OD.Quantity
FROM Sales.[Customer] AS C
  LEFT OUTER JOIN Sales.[Order] AS O
    ON C.CustomerId = O.CustomerId
  LEFT OUTER JOIN Sales.[OrderDetail] AS OD
    ON O.OrderId = OD.OrderId;
```

### Option 2: change join order

proposition: For each customer that has made an order (1 or more), obtain the product IDs and quantity of each product that makes up each order placed. Use an Inner Join first on the Order and Order Detail tables, and then use the results of that join to perform a Right Outer Join on the Customer table. 

- Note: this option also returns 2 results that are NULLs for the Customers that never placed an order. This is a result of the Right Outer join on the Customer table, which preserves the customers that never placed an order. 

```sql
Use Northwinds2022TSQLV7

SELECT C.CustomerId, O.OrderId, OD.ProductId, OD.Quantity
FROM Sales.[Order] AS O
  INNER JOIN Sales.[OrderDetail] AS OD
    ON O.OrderId = OD.OrderId
  RIGHT OUTER JOIN Sales.[Customer] AS C
     ON O.CustomerId = C.CustomerId
-- WHERE O.OrderID IS NULL
```

### Option 3: use parentheses

proposition: For each customer that has made an order (1 or more), obtain the product IDs and quantity of each product that makes up each order placed. Use a Left Outer Join on the Customer and the result of the Inner Join within the parenthesis. More exactly, pause the Left Outer Join to get the results of the Inner Join between the Order and Order Detail table, then use those results to complete the Left Outer Join with the Customer table.

- Note: this option also returns 2 results that are NULLs for the Customers that never placed an order. This is a result of the Left Outer join on the Customer table, which preserves the customers that never placed an order.

```sql
USE Northwinds2022TSQLV7

SELECT C.CustomerId, O.OrderId, OD.ProductId, OD.Quantity
FROM Sales.[Customer] AS C
  LEFT OUTER JOIN
      (Sales.[Order] AS O
         INNER JOIN Sales.[OrderDetail] AS OD
           ON O.orderid = OD.orderid)
    ON C.CustomerId = O.CustomerId;
```
